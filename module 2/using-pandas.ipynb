```python
city_id = dict()
for i in range(len(cities.city)):
    city_id[cities.id[i]] = cities.city[i]

state_id = dict()
for i in range(2):
    state_id[states.id[i]] = states.state[i]
```

```python
state_id
```

```python
city_id
```

```python
def recode_city(cid):
    return city_id[cid]

def recode_state(sid):
    return state_id[sid]
```

```python
yelp["city"] =  yelp.city_id.apply(recode_city)
```

```python
yelp["state"] = yelp.state_id.apply(recode_state)
```

```python
yelp.drop(["state_id", "city_id"], axis = 1)
```
