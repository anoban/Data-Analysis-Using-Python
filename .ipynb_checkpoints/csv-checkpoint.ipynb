```python
# there appears to be some years that are in text form :(

def unique(input_list):
    """
    Takes in a Python list objetct and returns only the unique elements
    """
    unique_elems = list()
    for elem in input_list:
        if elem not in unique_elems:
            unique_elems.append(elem)
        else:
            continue
    return unique_elems

unique(data_dict.get("Year"))
```

```python
def asNumeric(input_string):
    """
    Takes in a string as argument and returns a numeric value, either an integer or a float 
    depending on the input by type casting the string
    """
    if "." in input_string:
        try:
            floatDig = float(input_string)
        except ValueError:
            floatDig = None
        return floatDig
    elif "." not in input_string:
        try:
            intDig = int(input_string)
        except ValueError:
            intDig = None
        return intDig
```

```python
albums_with_numeric_years = [dictionary for dictionary in csv.DictReader(open("./Mod 1 Code & Data Files/albumlist.csv", "r")) if type(asNumeric(dictionary.get("Year"))) == int]
```

```python
albums_with_numeric_years
```

```python
sorted(unique([dictionary.get("Year") for dictionary in albums_with_numeric_years]))
```

```python
albums_years = dict()
for record in albums_with_numeric_years:
    albums_years[record.get("Album")] = int(record.get("Year"))
albums_years
```

```python
max(albums_years.values())
```

```python
for album, year in albums_years.items():
    if year >= 2000:
        print(album)
```

```python
year = 0
latest_album = 0

for record in albums_with_numeric_years:
    if int(record.get("Year")) > year:
        year = int(record.get("Year"))
        latest_album = record
```

```python
def calculate(str1, str2):
    calculated_number = 0
    # try block should be here!!!
    num1 = float(str1)
    num2 = float(str2)
    try:
        calculated_number = num1 / num2
    except ValueError:
        print ('Error!')
    return calculated_number

print(calculate('one', 'zero'))
```
