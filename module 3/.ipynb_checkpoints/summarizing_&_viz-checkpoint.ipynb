```python
yelp_data = pd.ExcelFile("yelp.xlsx").parse("yelp_data")
cities = pd.ExcelFile("yelp.xlsx").parse("cities")
states = pd.ExcelFile("yelp.xlsx").parse("states")
```

```python
yelp_data.columns, cities.columns, states.columns
```

```python
df = pd.merge(left = pd.merge(left = yelp_data, left_on = "city_id", right = cities, right_on = "id"), left_on = "state_id", right = states, right_on = "id")
df = df.drop(["city_id", "state_id", "id_x", "id_y"], axis = 1)
df.head(5)
```

```python
# .pivot_table() only considers numeric columns
pd.pivot_table(df, index = ["take_out"])
```

```python
# to use functions other than the default in .pivot_table() use the aggfunc parameter
# get the summation of reviews for all cities for each teke_out category
pd.pivot_table(df.loc[:, ["state", "city", "review_count"]], index = ["state", "city"], aggfunc = np.sum)
```

```python
pd.pivot_table(df, index = ["state", "city"], aggfunc = np.sum).loc[:, "review_count"]
```

```python
# get the maximums of review counts
pd.pivot_table(df.loc[:, ["state", "city", "review_count"]], index = ["state", "city"], aggfunc = max)
```

```python
df.columns
```

```python
# specifying separate aggfuncs for each columns
# summation for review counts and mean for stars :)
pd.pivot_table(df.loc[:, ['take_out', 'review_count', 'stars', 'city', 'state']], index = ["take_out", "state", "city"], 
               aggfunc = {"review_count": sum, "stars": np.mean})
```

# ***Data Visualization***

```python
pd.ExcelFile("yelp.xlsx").sheet_names
```

```python
pd.ExcelFile("yelp.xlsx").parse("yelp_data")
```

```python
pd.ExcelFile("yelp.xlsx").parse("cities")
```

```python
pd.ExcelFile("yelp.xlsx").parse("states")
```

```python
df = pd.merge(left = pd.merge(left = pd.ExcelFile("yelp.xlsx").parse("yelp_data"), left_on = "state_id",
                         right = pd.ExcelFile("yelp.xlsx").parse("states"), right_on = "id", how = "inner"),
         left_on = "city_id", right = pd.ExcelFile("yelp.xlsx").parse("cities"),
         right_on = "id").drop(["city_id", "state_id", "id_x", "state", "id_y"], axis = 1)
```

```python
df
```

```python
plt.figure(figsize = (20, 15))
plt.grid()
plt.rcParams["font.size"] = 20
plt.xlabel("$Stars$")
plt.ylabel("$Counts$")
for city in df.city.unique():
    plt.hist(df[df.city == city]["stars"], label = city, alpha = 0.6)
    plt.legend()
```

```python
plt.figure(figsize = (20, 10))
plt.subplot(1,2,1)
plt.grid()
plt.rcParams["font.size"] = 20
plt.xlabel("$Stars$")
plt.ylabel("$Counts$")
plt.hist(df[df.city == "Pittsburgh"]["stars"], color = "red", bins = 8)
plt.subplot(1,2,2)
plt.grid()
plt.rcParams["font.size"] = 20
plt.xlabel("$Stars$")
plt.ylabel("$Counts$")
plt.hist(df[df.city == "Las Vegas"]["stars"], color = "blue", bins = 8)
```

```python
plt.figure(figsize = (20, 10))
plt.grid()
plt.rcParams["font.size"] = 20
plt.hist([df[df.city == "Las Vegas"]["stars"], df[df.city == "Pittsburgh"]["stars"]], color = ["red", "green"], bins = 8, label = ["las Vegas", "Pittsburgh"])
plt.legend()
plt.xlabel("$Stars$")
plt.ylabel("$Counts$")
```

```python
plt.figure(figsize = (50, 70))
plt.grid()
plt.rcParams["font.size"] = 25
for category in df.category_0.unique():
    plt.plot(df[df.category_0 == category]["stars"], df[df.category_0 == category]["review_count"],
             marker = "o", markersize = 35, markeredgecolor = "black", linestyle = "none", label = category)
plt.legend()
plt.xlabel("$Stars$")
plt.ylabel("$Review count$")
```

```python
plt.figure(figsize = (10, 20))
plt.grid()
plt.plot([df[df.category_0 == category]["stars"] for category in df.category_0.unique()], df[df.category_0 == category]["review_count"],
             marker = "o", markersize = 15, markeredgecolor = "black", linestyle = "none")
plt.xlabel("$Stars$")
plt.ylabel("$Review count$")
```
